```python
from __future__ import print_function, division
import sys
import os
import numpy 
import pylab as pl
import pandas as pd

import os


%pylab inline

```

```python
dfmall = pd.read_excel("http://cosmo.nyu.edu/~fb55/PUI2016/data/pinc08_2_1_1.xls", header = 9) # all male
dfmwh = pd.read_excel("http://cosmo.nyu.edu/~fb55/PUI2016/data/pinc08_2_1_4.xls", header = 9) # white male
dfmbl = pd.read_excel("http://cosmo.nyu.edu/~fb55/PUI2016/data/pinc08_2_1_6.xls", header = 9) # black male
dfmhs = pd.read_excel("http://cosmo.nyu.edu/~fb55/PUI2016/data/pinc08_2_1_9.xls", header = 9) # hispanic male
dfmaz = pd.read_excel("http://cosmo.nyu.edu/~fb55/PUI2016/data/pinc08_2_1_8.xls", header = 9) # asian male

dffall = pd.read_excel("http://cosmo.nyu.edu/~fb55/PUI2016/data/pinc08_3_1_1.xls", header = 9) # all female
dffwh = pd.read_excel("http://cosmo.nyu.edu/~fb55/PUI2016/data/pinc08_3_1_4.xls", header = 9) # white female
dffbl = pd.read_excel("http://cosmo.nyu.edu/~fb55/PUI2016/data/pinc08_3_1_6.xls", header = 9) #black female
dffhs = pd.read_excel("http://cosmo.nyu.edu/~fb55/PUI2016/data/pinc08_3_1_9.xls", header = 9) #hispanic female
dffaz = pd.read_excel("http://cosmo.nyu.edu/~fb55/PUI2016/data/pinc08_3_1_8.xls", header = 9) #asian female
```

```python
allmales = {'all': dfmall, 'white': dfmwh, 'black': dfmbl, 'hispanic': dfmhs, 'asian': dfmaz}
allfemales = {'all': dffall, 'white': dffwh, 'black': dffbl, 'hispanic': dffhs, 'asian': dffaz}
```

```python

pl.scatter(allmales['white'][u'Median income'][1], allfemales['white'][u'Median income'][1],
           s = 50, color = '#F4D03F', label = 'White')
pl.scatter(allmales['black'][u'Median income'][1], allfemales['black'][u'Median income'][1],
           s = 50, color = '#FF8E00', label = 'Black')
pl.scatter(allmales['hispanic'][u'Median income'][1], allfemales['hispanic'][u'Median income'][
        1], s = 50, color = '#3498DB', label = 'Hispanic')
pl.scatter(allmales['asian'][u'Median income'][1], allfemales['asian'][u'Median income'][1],
           s = 50, color = '#2ECC71', label = 'Asian')



pl.xlim(0, 50000)
pl.ylim(0, 50000)

pl.title('Total Female Median Income vs Total Male Median Income by Race', fontsize = 12)
pl.ylabel('Female Median Income')
pl.xlabel('Male Median Income')
pl.legend(loc='best', fontsize = 'medium', scatterpoints = 1)
```

## Figure 1. Comparison of Total Female Median Income to Total Male Median Income by Race
