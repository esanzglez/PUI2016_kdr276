```python
from __future__ import print_function
import numpy as np
import pylab as pl


%pylab inline
```

## Chi-Square Distribution

```python
mysize = (2000 / (np.array(range(1, 100)))).astype(int)
```

```python
mysize
```

```python
chisqd = []
for c in range(0, len(mysize)):
    chisqd.append(np.random.chisquare(100, mysize[c]))
```

```python
chisqmeans = []
for c in range(0, len(mysize)):
    chisqmeans.append(chisqd[c].mean())
```

```python
pl.scatter(mysize, chisqmeans)
pl.xlabel('Sample Size')
pl.ylabel('Sample Mean')
pl.title('Chi Squared')
```

### Figure 1. Chi Squared Distribution Scatter Plot of Sample Means
##### The plot of the chi squared distribution approaches the shape of a normal distribution. Based on the Central Limit Theorem if we were to pull more than the 100 samples and run the same analysis a normal distribution shape would be even more evident

```python
pl.hist(chisqmeans, bins=50)
pl.title('Chi Squared')
pl.xlabel('Sample Mean')
```

### Figure 2. Chi Squared Distribution Histogram of Sample Means
#### The plot of the chi squared distribution approaches the shape of a normal distribution. Based on the Central Limit Theorem if we were to pull more than the 100 samples and run the same analysis a normal distribution shape would be even more evident
