Question: Is the new bus route improving commute time?

Null Hypothesis H0: the commuting time is longer than or the same with the new bus route as it was before

Alternative Hypothesis Ha: the commuting time is shorter with the new bus route than it was before

H0 = Tn >= To

Ha = Tn < To

Statistical Significance: p-value =.05

Information given: To mean = 36, standard deviation = 6
Sample size is 100

```python
z = (36 - 34.5)/ (6/(100**.5))
## z = population mean - sample mean / standard deviation/square root of the sample size
```

```python
print (z)
```

The results of the z-test show that there is a variance of 2.5 between the sample mean (34.5) and the population mean (36). Due to the fact that our statistical significance was choosen to be .05 and this result is futher than 2 sigmas away from the old mean, we can reject the null hypothesis.
